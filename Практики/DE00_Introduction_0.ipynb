В Python обратная косая черта — это специальный символ, применяемый для экранирования других символов.\
Например, \n обозначает символ новой строки, а \t — табуляцию. \

Хорошо: file_path = r"C:\Program Files\file.txt"

```python
A="C:\Program Files"
print(A)
```

```python
A=r"C:\Program Files"
print(A)
```

Целые числа (int) и числа с плавающей точкой (float) в Python

Целые числа (int):\
Представляют целые числа, положительные или отрицательные, без дробной части.\
Имеют произвольную длину (ограничено только памятью).\
Примеры: 5, -3, 1000, 0.

```python
a=2
print(type(a))
b=-45
print(type(b))
```

Числа с плавающей точкой (float):\
Представляют вещественные числа (с дробной частью).\
Имеют ограниченную точность (обычно 64 бита, что соответствует около 15-17 значащим цифрам).\
Примеры: 3.14, -2.5, 1.0, 2e-3 (что равно 0.002).

```python
a=2.0
print(type(a))
b=-45.0
print(type(b))
```

Важные моменты:\
Арифметические операции int и float, тоесть сложение, вычитание, умножение работают как ожидается:\
Смешанные операции: когда в выражении встречаются int и float, результат будет float.\


```python
a=6
b=3
c=a+b
print("Результат вычислений: ", c,', тип данных с =', type(c))
c=a-b
print("Результат вычислений: ", c,  ', тип данных с =', type(c))
c=a*b
print("Результат вычислений: ", c,', тип данных с =', type(c))
a=6.0
b=3
c=a+b
print("Результат вычислений: ", c,  ', тип данных с =', type(c))
c=a-b
print("Результат вычислений: ", c,  ', тип данных с =', type(c))
c=a*b
print("Результат вычислений: ", c,', тип данных с =', type(c))
```

### 2. Структуры данных

Списки(List) упорядоченные колекции элемнтов. Могут содержать данные любого типа 

```python
listExample=[1,2,3,4,5]
print(listExample)
```

```python
listExample2=['one','two','three','four','five']
print(listExample2)
```

Элементы можно добавлять, изменять или удалять. Они доступны по индексу.
Команда append добавляет элемент в конец списка. 

```python
listExample=[1,2,3,4,5]
listExample.append(6)
print(listExample)
```

```python
listExample2=['one','two','three','four','five']
listExample2.append('six')
print(listExample2)
```

Метод append принимает только 1 аргумент. Если задать несколько появится ошибка 

```python
listExample=[1,2,3,4,5]
listExample.append(6,7)
print(listExample)
```

Можно добавить 1 элемент списко в другой. ОДнако метод не распоковывает вложенные списки, а добавляет их единым элементом. 

```python
list1=[1,2,3]
list2=[4,5]
list1.append(list2)
print(list1)
```

Для соедениения двух списков необходимо использовать метод extend 

```python
list1=[1,2,3]
list2=[4,5]
list1.extend(list2)
print(list1)
```

Существует несколько других способов добавления элементов в список. Например, можно использовать оператор +

```python
list1=[1,2,3]
list2=[4,5]
list1+=list2
print(list1)
```

### 3. Условные операторы

В Python используются условные операторы if, elif и else. Они позволяют выполнять код в зависимости от выполнения определённых условий.
if Проверяет одно условие и выполняет соответствующий блок кода, если оно истинно. 
elif Проверяет несколько условий. Если условие if ложно, проверяется условие elif. 
else Выполняет блок кода, если все предыдущие условия ложны. Позволяет указать, что должно произойти, если ни одно из предыдущих условий не выполнено.

```python
numTaskStud=0 #записутите блок несколько раз изменяя данное условие
numTask=10
if numTaskStud == 0:
    print('Студент не приступил к занятиям')

elif numTaskStud >= numTask:
    if numTaskStud == numTask:
        print('Студент выполнил основную программу')
    else:
        print('Судент перевыполнил программу')
else:
    print('Студент не готов к сессии')
```

### 4. Циклы

В Python есть два основных типа циклов: цикл for и цикл while. Они позволяют многократно выполнять одни и те же действия в зависимости от заранее определённых условий. 
Цикл for
Используется для перебора элементов в последовательности (например, в списке или строке) и выполнения одних и тех же действий для каждого элемента. 

```python
for i in range(5): print(i)
```

```python
students = ["Иванов", "Петров", "Сидоров","Смирнов"] 
for students in students: print(students)
```

Цикл while
Выполняет блок кода, пока заданное условие остаётся истинным.
Цикл продолжает выполняться до тех пор, пока условие остаётся истинным (True), как только условие становится ложным (False), выполнение цикла прекращается. 


```python
count = 1 
while count <= 5: 
    print(count) 
    count += 1
```

### 3. Функции

Функции в Python — это блоки кода, которые выполняют определённую задачу. Они могут принимать входные данные (аргументы), выполнять действия и возвращать результат. В Python функции создаются с помощью ключевого слова def. Ниже приведены примеры функций разных типов: без аргументов, с аргументами, с возвращаемым значением и встроенных. 

```python
def say_hello(): 
    print("Привет, мир!")  

say_hello()
```

```python
def say_hello(Name): 
    print("Привет, "+Name)  

say_hello('Студент')
```
