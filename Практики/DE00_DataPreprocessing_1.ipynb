Важно правильно подобрать графический тип представления данных для различных типов данных. 
Рассмотрим неудачные примеры использования графических форм представления данных на двух искусственно сгенерированных наборов данных - графика синуса и набора "две луны" 
Сформируем массив точек для каждого набора данных и сохраним его в наиболее удобном формате хранения данных - .csv

```python
def GetSin(duration,sample_rate,frequency,amplitude):
    t = np.linspace(0, duration, int(sample_rate * duration), endpoint=True)
    sine_wave = amplitude * np.sin(2 * np.pi * frequency * t)
    return sine_wave,t

# зададим 2 типа коэффициентов для формирования двух разных наборов синуса. 
# Один синус будет относиться к первому класс данных, второй - ко второму.     
# Так как при анализе длительность и количество точек должны сопадать у обоих наборов,
#менять будем только частоту и амплитуду 
#задаем параметры
duration = 3  # длительность (например в секундах)
sample_rate = 100 # количество точек (количество точек в секунде)
frequency1 = 1  # Частота
amplitude1 = 1 #Апплитуда
frequency2 = 4  
amplitude2 = 2

# Генерируем синус
sin_wave1,t1 = GetSin (duration,sample_rate,frequency1,amplitude1)
sin_wave2,t2 = GetSin (duration,sample_rate,frequency2,amplitude2)
# посмотрим размеры получившихся массивов
print(len(sin_wave1))
print(len(sin_wave2))
print(len(t1))
print(len(t2))
#print(sin_wave1.shape)
```

Создадим DataFrame из полученных значений. При этом первый синус у нас будет относиться к первому классу, а второй ко второму 

```python
#создадим пустой DataFrame
df = pd.DataFrame(columns=['x_coord', 'y_coord'])
# заполним пустой датафрейм значениями первого синуса и меткой класса 
df['x_coord']=t1
df['y_coord']=sin_wave1
df['class_label'] = 1
#выведем первые и последние 5 строк полученного датафрейма для проверки, а также посмотрим размер датафрейма 
print(df.head(5))
print(df.tail(5))
print(df.shape)
```

```python
#Проведем такую же процедуру для массива синуса 2 и объеденим их в общим датафрейм
df2 = pd.DataFrame(columns=['x_coord', 'y_coord'])
# заполним пустой датафрейм значениями первого синуса и меткой класса 
df2['x_coord']=t2
df2['y_coord']=sin_wave2
df2['class_label'] = 2
print(df2.head(5))
print(df2.tail(5))
print(df2.shape)
```

```python
# объеденим df1 и df2
result_sin = pd.concat([df, df2], axis=0, ignore_index=True)
print(result_sin.head(5))
print(result_sin.tail(5))
print(result_sin.shape)
```
